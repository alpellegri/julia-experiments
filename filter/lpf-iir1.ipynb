$$ y[n] = b_0x[n] + a_1y[n-1] $$

where,
$$ b_0 = b_1 $$
$$ b_0 + b_1 + a_1 = 1 $$

therefore,
$$ a_1 = \alpha $$
$$ b_0 = {1-\alpha} $$


$$ H(z) = \frac{1-\alpha}{1-\alpha z^{-1}} $$

$$ H(e^{j\omega}) = \frac{1-\alpha}{1-\alpha e^{-j\omega}} $$

$$ |H(j\omega)|^2 = \frac{1-2\alpha+\alpha^2}{(1-\alpha e^{-j\omega}){(1-\alpha e^{-j\omega})}^*} $$

using Eulero formula:

$$ e^{j\omega} = cos\omega + jsin\omega $$

therefore:

$$ |H(j\omega)|^2 = \frac{1-2\alpha+\alpha^2}{(1-\alpha cos\omega-j\alpha sin\omega)(1-\alpha cos\omega+j\alpha sin\omega)} $$

$$ |H(j\omega)|^2 = \frac{1-2\alpha+\alpha^2}{1-2\alpha cos\omega+\alpha cos^2\omega} $$

cut-off frequency:

$$ \frac{1}{2} = \frac{1-2\alpha+\alpha^2}{1+\alpha^2-2\alpha cos\omega} $$

$$ 1+\alpha^2-2\alpha cos\omega = 2-4\alpha+2\alpha^2 $$

$$ \alpha^2 - (4-2cos\omega)\alpha + 1 = 0 $$

solving the quadratic equation in $\alpha$

recall the roots of a quadratic equation:

$$ ax^2+bx+c=0 $$

has roots on:

$$ x = \frac{-b\pm\sqrt{b^2-4ac}}{2a} $$

therefore:

$$ \alpha = \frac{2(2-cos\omega)\pm\sqrt{(4-2cos\omega)^2-4}}{2} $$

$$ \alpha = \frac{2(2-cos\omega)\pm\sqrt{12+4cos^2\omega -16cos\omega}}{2} $$



first order IIR filter:

$$ y[n] = b_0x[n] + b_1x[n-1] + a_0y[n-1] $$

where,
$$ b_0 = b_1 $$
$$ b_0 + b_1 + a_0 = 1 $$

therefore,
$$ a_0 = \alpha $$
$$ b_0 = b_1 = \frac{1-\alpha}{2} $$

$$ H(z) = {\frac{1-\alpha}{2}}{\frac{1+z^{-1}}{1-\alpha z^{-1}}} $$

substituting
$$ z = e^{j\omega} $$

$$ H(e^{j\omega}) = {\frac{1-\alpha}{2}}{\frac{1+e^{-j\omega}}{1-\alpha e^{-j\omega}}} $$

$$ |H(e^{j\omega})|^2 = {\frac{(1-\alpha)^2}{2^2}} {\frac{(1+e^{-j\omega})(1+e^{-j\omega})^*}{(1-\alpha e^{-j\omega}){(1-\alpha e^{-j\omega})}^*}} $$

using Eulero formula:

$$ e^{j\omega} = cos\omega + jsin\omega $$

therefore:

$$ |H(j\omega)|^2 = {\frac{(1-\alpha)^2}{4}} \frac{(1+cos\omega-jsin\omega)(1+cos\omega+jsin\omega)}{(1-\alpha cos\omega-j\alpha sin\omega)(1-\alpha cos\omega+j\alpha sin\omega)} $$

solving numerator:

$$ (1+cos\omega-jsin\omega)(1+cos\omega+jsin\omega) = 1 + 2cos\omega + cos^2\omega + sin^2\omega - jsin\omega(1+cos\omega) + (1+cos\omega)jsin\omega $$

$$ cos^2\omega + sin^2\omega = 1 $$

$$ (1+cos\omega-jsin\omega)(1+cos\omega+jsin\omega) = 2(1+cos\omega) $$

solving denominator:

$$ (1-\alpha cos\omega-j\alpha sin\omega)(1-\alpha cos\omega+j\alpha sin\omega) = 1 - 2\alpha cos\omega + \alpha^2cos^2\omega - \alpha^2sin^2\omega + j\alpha sin\omega(1+\alpha cos\omega) + (1+\alpha cos\omega)j\alpha sin\omega $$

$$ (1-\alpha cos\omega-j\alpha sin\omega)(1-\alpha cos\omega+j\alpha sin\omega) = 1 - 2\alpha cos\omega + \alpha^2 $$

$$ |H(j\omega)|^2 = {\frac{(1-\alpha)^2}{4}} \frac{2(1+cos\omega)}{1 - 2\alpha cos\omega + \alpha^2} $$

$$ |H(j\omega)|^2 = {\frac{(1-\alpha)^2}{2}} \frac{1+cos\omega}{1 - 2\alpha cos\omega + \alpha^2} $$

cut-off frequency:

$$ \frac{1}{2} = |H(j\omega)|^2 $$

$$ \frac{1}{2} = {\frac{(1-\alpha)^2}{2}} \frac{1+cos\omega}{1 - 2\alpha cos\omega + \alpha^2} $$

$$ 1 = (1-\alpha)^2 \frac{1+cos\omega}{1 - 2\alpha cos\omega + \alpha^2} $$

$$ 1 - 2\alpha cos\omega + \alpha^2 = (1-2\alpha+\alpha^2)(1 + cos\omega) $$

$$ 1 - 2\alpha cos\omega + \alpha^2 = 1+cos\omega - 2\alpha - 2\alpha cos\omega +\alpha^2 + \alpha^2 cos\omega $$

$$ 0 = cos\omega - 2\alpha + \alpha^2 cos\omega $$

solving the quadratic equation in $\alpha$

recall the roots of a quadratic equation:

$$ ax^2+bx+c=0 $$

has roots on:

$$ x = \frac{-b\pm\sqrt{b^2-4ac}}{2a} $$

therefore:

$$ \alpha = \frac{2\pm\sqrt{4-4cos^2\omega}}{2cos\omega} $$

$$ \alpha = \frac{2\pm\sqrt{4sin^2\omega}}{2cos\omega} $$

$$ \alpha = \frac{1\pm{sin\omega}}{cos\omega} $$

$$ \alpha = \frac{1-{sin\omega}}{cos\omega} $$
