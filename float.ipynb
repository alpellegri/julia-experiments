```julia
mutable struct cFloat
    s::Int64
    e::Int64
    m::Int64
    function cFloat(s, e, m)
        new(s, e, m)
    end
end
```

```julia
a = cFloat(1,0,500)
```

```julia
function encode(v)
    N = 9
    s = sign(v)
    e = ceil(log2(abs(v)))
    e = max(e, 5)
    m = abs(v)*2^(N-e)
    m = floor(m)
    m = m/2^(N)
    s, e, m
end

function decode(s, e, m)
    v = s*2^e*m
end
```

```julia
s, e, m = encode(10.35)
```

```julia
decode(s,e,m)
```

```julia
x = -100.35
s, e, m = encode(x)
y = decode(s,e,m)
x-y
```

```julia
x = 1:.1:200
y = zeros(length(x))
for i=1:length(x)
    s, e, m = encode(x[i])
    y[i] = decode(s,e,m)
end
```

```julia
plot([x x],[100*(x-y) 100*x*.005])
```
