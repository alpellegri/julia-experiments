```julia
using Plots
import GR
using IJulia
using LinearAlgebra
```

```julia
DIM = 2
N = 8

Rot(θ) = [cos(θ) -sin(θ); sin(θ) cos(θ)];
update(m, θ, d) = (m*Rot(θ) - [d*ones(N) zeros(N)])

model = 2 .* (rand(N, DIM) .- .5)
# model = [0 0; 0 1; 1 0]
```

```julia
scatter(model[:,1], model[:,2])
plot!(lims = (-3, 3), size = (200,200), legend = false)
```

```julia
P = 1.5
θ = π/16
```

```julia
myview = update(model, θ, P)
```

```julia
scatter(model[:,1], model[:,2])
scatter!(myview[:,1], myview[:,2])
```

```julia
for ii in 1:50
    myview = update(myview, θ, 0)
    scatter(myview[:,1], myview[:,2])
    plot!(lims = (-3, 3), size = (200,200), legend = false) |> IJulia.display
    IJulia.clear_output(true)
    sleep(0.2)
end
```
