```julia
using DifferentialEquations

p = (0.230605, 0.0443577)
N = 10e6
C = 172
f = function (du, u, p, t) # Define f as an in-place update into du
    N = u[1]+u[2]+u[3]
    b, g = p
    du[1] = -b*u[1]*u[2]/N
    du[2] = b*u[1]*u[2]/N - g*u[2]
    du[3] = g*u[2]
end
```

```julia
u0 = [N-C; C; 0.0]; tspan = (0.0, 50.0)
prob = ODEProblem(f, u0, tspan, p);
# Solve the problem
sol = solve(prob);
```

```julia
using Plots; gr() # Using the Plotly Backend
plot(sol,title="All Plots.jl Attributes are Available")
```

```julia
using DifferentialEquations
using ParameterizedFunctions, LinearAlgebra

N = 10e6
C = 172

f = @ode_def LotkaVolterraTest begin
    dS = -β*S*I/N
    dI = β*S*I/N - γ*I
    dR = γ*I
end β γ N
```

```julia
u0 = [N-C; C; 0.0]                      # initial values
p = [0.230605, 0.0443577, N]               # parameters used, these need to be estimated from the data
tspan = (0.0, 50.0)                     # sample of 3000 observations over the (0,30) timespan
prob = ODEProblem(f, u0, tspan,p)

sol = solve(prob,Tsit5(),reltol=1e-9,abstol=1e-9);
```

```julia
using Plots; gr() # Using the Plotly Backend
plot(sol,title="All Plots.jl Attributes are Available")
```
