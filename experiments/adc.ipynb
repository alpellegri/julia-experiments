```julia
using PyPlot

x = 0:.01:1000
y=0

LIM = (2^10-1)

y = zeros(length(x))
for i=1:length(x)
    v = x[i]*512
    while v>LIM
        v = v/8
    end
    y[i]=v
end

plot((x),y, color="blue")
gcf()
# plot(y_18,x,color="red")
# gcf()
# plot(y_36,x,color="red")
# gcf()
# plot(y_54,x,color="red")
# gcf()
# plot(x,y,color="green")
;
```

```julia
plot(log(x),log(y), color="blue")
gcf()
```

> *float24:* $$f = m 2^e$$

> *RSS* $$RSS = log2(f)$$

> *RSS* $$RSS = elog2(m)$$

> $$RSS(d) = log(k) - \frac{1}{n}log(d)$$

> $$RSS(d) = A - \frac{1}{n}log(d)$$ where $A = log(k)$

> **Trick:** to avoid infinity change $f$ (a sort of saturation) into:

> $$f(x,y) = \frac{k}{x^2+y^2 + e}$$

> $$RSS(x,y) = A - {\frac{n}{2}}log(x^2+y^2 + e)$$
