```julia
using Random, Distributions
using Plots
```

```julia
N = 100000

x = rand(Uniform(-1,0), N)
y = rand(Uniform(0,1), N)
z=(x.+y)/2
histogram(x, nbin=100)
histogram!(y, nbin=100)
histogram!(z, nbin=100)
```

```julia
N = 100000

x = rand(Normal(-1., .5), N)
y = rand(Normal(1., .5), N)
z=(x.+y)/2
histogram([x,y,z], nbin=100, layout = (3,1), xlims=(-3,3))
```

```julia
N = 100000

f(x,y) = (y./(x.+y))

x = rand(Normal(9., .5), N)
y = rand(Normal(1., .5), N)
z = f(x,y)
histogram([x,y,z], nbin=100, layout = (3,1))
```
